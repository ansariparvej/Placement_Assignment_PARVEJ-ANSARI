```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
# Music features Dataset:
df = pd.read_csv('useful/Q7.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
df['label'].unique()
```

```python
# Split X andy:

X = df.drop(['label', 'filename'], axis=1)
y = df['label']
X.shape, y.shape
```

```python
X.describe().T
```

```python
for column in X.columns:
    # Extract the minimum and maximum values from the column
    min_value = X[column].min()
    max_value = X[column].max()
    
    # Print the minimum and maximum values for the column
    print("Column:", column)
    print("Min Value:", min_value)
    print("Max Value:", max_value)
    print()
```

```python
# Column wise unique values:
for col in X:
    print(f'{col}: {X[col].unique()}')
```

```python
y.unique()
```

```python
# train-test-split:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)


# Applying Scaling
scaler = MinMaxScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

# Handling missing values
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train_s)
X_test = imputer.transform(X_test_s)


# Training using Unsupervised algorithm:
model = KNeighborsClassifier(n_neighbors=6)
model.fit(X_train, y_train)

# Testing:
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print('Accuracy of model: ', acc*100, '%')
                                                                                                       
```

```python
# Let's predict:
Input = X_test[50]
y_pred = model.predict([Input])
print("Music Genre ==> ",y_pred[0])
```
